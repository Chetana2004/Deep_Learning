```python
# a. Import libraries
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# b. Load and preprocess dataset
data = pd.read_csv("creditcard.csv")
X = data.drop("Class", axis=1)
y = data["Class"]
X = StandardScaler().fit_transform(X)
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)
```

```python
# c. Encoder
inp = Input(shape=(X.shape[1],))
enc = Dense(16, activation='relu')(inp)
enc = Dense(8, activation='relu')(enc)
```

```python
# d. Decoder
dec = Dense(16, activation='relu')(enc)
out = Dense(X.shape[1], activation='sigmoid')(dec)

autoencoder = Model(inp, out)
```

```python
# e. Compile & train
autoencoder.compile(optimizer=Adam(0.001), loss='mse', metrics=['mae'])
autoencoder.fit(X_train, X_train, epochs=5, batch_size=256, validation_split=0.1, verbose=1)
```

```python
# Evaluate reconstruction loss
loss, mae = autoencoder.evaluate(X_test, X_test)
print(f"Test Loss: {loss:.5f}, MAE: {mae:.5f}")
```
